```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "College.csv"  # bringing dataset to load
college = pd.read_csv(r"C:\Users\kavya\Downloads\College.csv")

college.head()  ## Display first few rows
```

```python
pd.set_option('display.max_rows', None)  # Allows displaying all rows
print(college)  # Shows all rows
```

```python
college = college.rename({'Unnamed: 0': 'College'}, axis=1) # assign meaningful label to first column
college = college.set_index('College')

college.head()  # display updated dataset
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']])    # Scatterplot matrix of selected columns
plt.show()
```

```python
# Boxplot of Outstate tuition grouped by Private/Public status
plt.figure(figsize=(8,6))
sns.boxplot(x=college["Private"], y=college["Outstate"])
plt.title("Outstate Tuition vs Private/Public")
plt.show()
```

```python
# Create Elite variable
college['Elite'] = pd.cut(college['Top10perc'], bins=[0, 50, 100], labels=['No', 'Yes'])
# Count the number of Elite universities
college['Elite'].value_counts()
```

```python
college['Elite'] = pd.cut(college['Top10perc'],
[0,0.5,1],
labels=['No', 'Yes'])
college['Elite'].value_counts()
```

```python
# Boxplot of Outstate tuition grouped by Elite status
plt.figure(figsize=(8,6))
sns.boxplot(x=college["Elite"], y=college["Outstate"])
plt.title("Outstate Tuition vs Elite Status")
plt.show()
```

```python
# Histograms of selected quantitative variables
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

columns = ['Apps', 'Enroll', 'PhD', 'Grad.Rate']
for i, col in enumerate(columns):
    ax = axes[i//2, i%2]
    ax.hist(college[col], bins=20, edgecolor='black')
    ax.set_title(f"Histogram of {col}")

plt.tight_layout()
plt.show()
```

### Data Exploration summary

1. Tuition Trends: Private institutions generally charge a premium for out-of-state students, making affordability a key differentiator from public universities.
2. Elite vs. Standard: Schools where over half of the incoming class hails from the academic elite (top 10% of high school) tend to have steeper tuition fees, possibly reflecting exclusivity and prestige.
3. Application Disparity: While some universities are flooded with applications, others attract a more selective or niche pool of candidates, highlighting vast differences in popularity and outreach.
4. Classroom Experience: Student-to-faculty ratios vary significantly, hinting at diverse educational environments—from intimate, discussion-driven classes to large lecture halls with hundreds of students.
